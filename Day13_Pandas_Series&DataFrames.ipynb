```python
#Pandas DataFrame
#Create from Dictionary
data = {
    'Name': ['Aisha', 'Tunde', 'Fatima'],
    'Age': [25, 30, 22],
    'City': ['Abuja', 'Lagos', 'Kano']
}
df = pd.DataFrame(data)
print(df)

#Accessing data
print(df['Name'])         
print(df.loc[0])          
print(df.iloc[0])         

#Useful methods
print(df.head())          # First 5 rows
print(df.tail())          # Last 5 rows
print(df.info())          # Summary info
print(df.describe())      # Stats summary
```
