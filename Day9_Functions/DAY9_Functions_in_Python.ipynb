```python
#Defining a function
def say_hello():
    print("Hello!")
#Calling the function
say_hello()

#Parameters and Arguments
def greet(name):
    print("Hello,", name)
greet("Aminah")
#name is a parameter
#Aminah is an argument

#Function can also give back a value using return
def square(num):
    return num * num
result = square(4)
print(result)

#If you dont pass anything python can use default
def greet(name="Data Learner"):
    print("Hello,", name)
greet()             
greet("Aminah")     

#*args and **kwargs; help when you donâ€™t know how many arguments someone will pass
def add_all(*numbers):
    return sum(numbers)
print(add_all(1, 2, 3, 4))

#*args = Multiple Positional Arguments (pack extra arguments into a tuple)
def add_numbers(*args):
    print(args)
    return sum(args)
print(add_numbers(1, 2, 3))  

def print_all_names(*names):
    for name in names:
        print("Hello", name)
print_all_names("Aminah", "Fatima", "Zainab")

#**kwargs = Multiple Keyword Arguments (pack extra named arguments into a dictionary)
def print_info(**kwargs):
    print(kwargs)
print_info(name="Aminah", age=20, course="Data Science")

#You can combine them but arrangement matters
def describe_person(*args, **kwargs):
    print("Positional:", args)
    print("Keyword:", kwargs)
describe_person("Student", "Nigeria", name="Aminah", age=20)
```

```python
#Practice Exercises
#Create a function greet_user() that prints "Hello, user!"
def greet_user():
    print("Hello, user!")
greet_user()

#Create a function greet(name) that prints "Hello, name!"
def greet(name):
    print(f"Hello, {name}!")
greet("Aisha")

#Create a function grade(score) that:
#Prints "A" if score >= 90
#Prints "B" if score >= 80
#Prints "C" if score >= 70
#Else, prints "Fail"
def grade(score):
    if score >= 90:
        print("A")
    elif score >= 80:
        print("B")
    elif score >= 70:
        print("C")
    else:
        print("Fail")
grade(44)

#Write a function add_all(*numbers) that returns the sum of all the numbers passed
def add_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total
print(add_all(1,3,5,7))

#Write a function show_info(**kwargs) that prints each key and value in the form key: value
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
show_info(name="Aminah", course="Data Science", country="Nigeria")

# Write a function full_report(*args, **kwargs) that:
#Prints the sum of all positional arguments.
#Then prints all keyword data.
def full_report(*args, **kwargs):
    print("Total Score:", sum(args))
    print("Details:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
full_report(70, 80, 90, name="Aminah", course="Python", status="Passed")
```
